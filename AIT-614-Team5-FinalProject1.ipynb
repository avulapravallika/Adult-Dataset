```
df_sel = df.select('Country','Territorial emissions')
display(df_sel.take(5))
```

```
display(df.groupBy('Country').agg({'Territorial emissions': 'mean'}))
```

```
from pyspark.sql.functions import sum 
df_sel = df.groupBy('Country').agg(sum('Territorial emissions').alias('total_emissions'))
display(df_sel.sort('total_emissions', ascending=False))
```

```
from pyspark.sql.functions import sum 
df_sel = df.groupBy('LA CO2 Sector','Country').agg(sum('Territorial emissions').alias('total_emissions'))
display(df_sel.sort('total_emissions', ascending=False))
```

```
from pyspark.sql.functions import sum 
df_sel = df.groupBy('LA CO2 Sector','LA CO2 Sub-sector').agg(sum('Territorial emissions').alias('total_emissions'))
display(df_sel.sort('total_emissions', ascending=False))
```

```
from pyspark.sql.functions import sum 
df_sel = df.groupBy('Calendar Year','Country').agg(sum('Territorial emissions').alias('total_emissions'))
display(df_sel.sort('total_emissions', ascending=False))
```
